```python
from datasets import load_from_disk
import matplotlib.pyplot as plt
import random
```

```python
def display_sample(dataset, n=5):
    idx = random.sample(range(len(dataset)), n)
    for i in idx:
        img = dataset[i]['image']
        plt.figure(figsize=(5, 5))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Caption: {dataset[i]['caption']}")
        plt.show()
display_sample(ds, 5)
```
